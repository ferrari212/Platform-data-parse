# Transversal and longitudinal ground speed

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('raw_data/IMO9371361_201810_SeapathGPSVbw.csv', header=0, sep=";")
```

```python
print(list(data.columns.values))
```

```python
display(data[data["TransGroundSpeed"].isnull().values])
```

```python
df = pd.DataFrame(data)
```

```python
# Filtering NaN lines
df['TransGroundSpeed'] = pd.to_numeric(df['TransGroundSpeed'], errors='coerce')
df = df.dropna(subset=['TransGroundSpeed'])
# df['Longitude'] = df['Longitude'].astype(int)
```

```python
display(df[df["TransGroundSpeed"].isnull().values])
```

```python
# Create a latitude and longitude location
loc = pd.DataFrame({
    "uid": df["uid"].values,
    "TransGroundSpeed": df["TransGroundSpeed"].values,
    "LonGroundSpeed": df["LonGroundSpeed"].values
})
```

```python
display(loc.head())
```

```python
loc.to_csv('output-data/GroundSpeed.csv', sep=";")
```

```python
# definitions for the axes
left, width = 0.1, 0.9
bottom, height =  0.1, 0.65

rect_scatter = [left, bottom, width, height]
```

```python
# start with a rectangular Figure
fig = plt.figure(1, figsize=(10,10))
axScatter = plt.axes(rect_scatter)

# the scatter plot:
axScatter.plot(loc.LonGroundSpeed)
axScatter.grid()
plt.show()
```

```python
# start with a rectangular Figure
fig = plt.figure(1, figsize=(10,10))
axScatter = plt.axes(rect_scatter)

# the scatter plot:
axScatter.plot(loc.TransGroundSpeed)
axScatter.grid()
plt.show()
```

```python
# start with a rectangular Figure
fig = plt.figure(1, figsize=(10,10))
axScatter = plt.axes(rect_scatter)

# the scatter plot:
axScatter.plot(loc.LonGroundSpeed, loc.TransGroundSpeed)
axScatter.set_xlabel("Longitudinal Ground Speed (m/s)")
axScatter.set_ylabel("Transversal Ground Speed (m/s)")
axScatter.grid()
plt.show()
```
